```python
import xml.etree.ElementTree as ET
import pickle
import os
from os import listdir, getcwd
from os.path import join
```

```python
# get all class tag
def gen_classes(image_id):
    classes = []
    in_file = open('/Users/jialeichen/Downloads/%s.xml'%(image_id))
    tree = ET.parse(in_file)
    root = tree.getroot()    
    for obj in root.iter('object'):
        cls_name = obj.find('name').text
        if cls_name in classes:
            pass
        else:
            classes.append(cls_name)
    return classes, len(classes)
```

```python
classes, class_len = gen_classes('test')
print(classes)
print(class_len)
```

```python
image_width = 640.0
image_height = 480.0
def convert_annotation(identity_num, path, image_id):
    in_file = open('%s/%s.xml'%(path, image_id))
    if not os.path.exists('%s'%(path)+'/labels/'+'%s'%image_id):
        os.makedirs('%s'%(path)+'/labels/'+'%s'%image_id)
    
    tree = ET.parse(in_file)
    root = tree.getroot()
    for obj in root.iter('object'):
        box_name = obj.find('name').text
        for polygon in obj.iter('polygon'):
            img_id = polygon.find('t').text
#             print(box_name + ' ' + img_id)
            xs = []
            ys = []
            for pt in polygon.iter('pt'):
                x = pt.find('x').text
                y = pt.find('y').text
                xs.append(x)
                ys.append(y)
#                 print(str(id) + ' ' + box_name + ' ' + img_id + ' ' + x + ' ' + y)
            width = float(xs[2]) - float(xs[0])
            height = float(ys[2]) - float(ys[0])
#             print(xs)
#             print(ys)
#             print(width)
#             print(height)
            center_x = float(xs[0]) + width / 2.0
            center_y = float(ys[0]) + height / 2.0
#             print(center_x)
#             print(center_y)
            
            width_norm = width / image_width
            height_norm = width / image_height
            center_x_norm = center_x / image_width
            center_y_norm = center_y / image_height
#             print(width_norm)
#             print(height_norm)
#             print(center_x_norm)
#             print(center_y_norm)
            
            new_line = str(0) + ' ' + str(identity_num) + ' ' + '%.6f'%center_x_norm + ' ' + '%.6f'%center_y_norm + ' ' + '%.6f'%width_norm + ' ' + '%.6f'%center_y_norm
#             print(new_line)
            
#             print(image_id + ' ' + image_id[0])
            out_file = open('%s'%(path)+'/labels/%s/%s.txt'%(image_id, img_id.zfill(6)), 'a')
            out_file.write(new_line + '\n')
        identity_num += 1
    return identity_num
# convert_annotation('/Users/jialeichen/Documents/NYU/2022Spring/Image\ and\ Video\ Processing/PROJECT/Dataset/', '1. Lapse Clip output')

```

```python
identity_cnt = 0
xml_file_name = ['0001', '0002', '0003', '0004', '0005', '0006', '0007', '0008']
for xml_file in xml_file_name:
    identity_num = convert_annotation(identity_cnt, '/Users/jialeichen/Documents/NYU/2022Spring/Image_and_Video_Processing/PROJECT/Dataset/annotations', xml_file)
#     print(identity_num)
    identity_cnt = identity_num
```
